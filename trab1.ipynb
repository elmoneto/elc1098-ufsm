```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df_temp = pd.read_csv("dados.csv",usecols=[0,1,2],encoding="latin_1")
```

```python
nomes_jogadores = ['Ágata','Alonso','Bárbara','Marta','Roberto','Romário','Ronaldo','Shelda'];
```

```python
with open('dados.csv', newline='',encoding="latin_1") as f:
    leitor = csv.reader(f)
    lista = list(leitor)
```

```python
# Remoção de itens desnecessários para a análise
for item in lista:
    del item[3:-1]     
    
for item in lista:
    del item[-1]
```

```python
#Correção de codificação de caracteres, remoção de espaços em branco e capitalização
#Divisão de string com nomes de jogadores -> strings cada uma com um nome de jogador
for item in lista:
    jogadores = item[1].split(',')
    item[1] = []
    for jogador in jogadores:
        jogador = jogador.replace(" ","")
        jogador = jogador.capitalize()
        if('gata' in jogador):
            jogador = 'Ágata'
        if('bara' in jogador):
            jogador = 'Bárbara'
        if('erto' in jogador):
            jogador = 'Roberto'
        if('mario' in jogador):
            jogador = 'Romário'
        item[1].append(jogador)
    
    item[2] = item[2].capitalize()
```

```python
lista
lista_para_df = []
```

```python
#Criação de dataframe que preenche colunas de jogadores com 1 ou 0 (True ou False),
#dependendo da presença ou ausência do jogador na partida
for i in range(1,len(lista)):
    lista_temp = []
    lista_temp.append(lista[i][0])
    for jogador in nomes_jogadores:
        if jogador in lista[i][1]:
            lista_temp.append(1)
        else:
            lista_temp.append(0)
    if(lista[i][2] == 'Ganhou'):
        lista_temp.append(1)
    else:
        lista_temp.append(0)
    lista_para_df.append(lista_temp)
```

```python
colunas = ["Partida",*nomes_jogadores,"Resultado"]
df = pd.DataFrame(lista_para_df,columns=colunas)
```

```python
df
```

```python
df = df.drop(['Partida'],axis=1)
```

```python
df
```

```python
suporte = apriori(df, min_support=0.1,use_colnames=True)
```

```python
suporte
```

```python
regras = association_rules(suporte, metric="confidence", min_threshold=0.1)
```

```python
regras
```

```python
regras_sort = regras.sort_values(by=['confidence'],ascending=False)
```
